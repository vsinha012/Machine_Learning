#Contatenation and Merging

```python
import pandas as pd
import numpy as np
```

```python
### Concatenation
df1 = pd.DataFrame({'A': [1,2,3,4], 'B': ['ab', 'ban', 'run', 'del']})
df1
```

```python
df2 = pd.DataFrame({'A': [23,45,67,89], 'D': ['runn', 'catch', 'ball', 'del']})
df2
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1,df2], ignore_index=True)
```

```python
## Merge
pd.merge(df1, df2, on='A', how='right')
```

```python
# Since there is no common value, if you just merge df1 and df2, there is no common A value between the two 
# data frames to merge.  Hence the result is no record. 
pd.merge(df1, df2)
```

```python
# You use how parameter to specify how you want to merge 
# the two data frame i.e. inner, outer, left join, right join
pd.merge(df1, df2, on='A', how='outer')
```

```python
pd.merge(df1, df2, on='A', how='left')
```

```python
pd.merge(df1, df2, on='A', how='right')
```
