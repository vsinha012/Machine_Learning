```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


%matplotlib inline
```

```python
df = pd.read_csv('shopping_data.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
data = df.iloc[:, 3:5].values
```

```python
# To know the clusters that we want out data to be split to. We will again use the scipy library to create the dendograms for our dataset. 
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(28,12))
plt.title("Customer Dendograms")
dend = shc.dendrogram(shc.linkage(data, method='ward'))
```

```python
# To create an instance of Agglomerative Clustering using the euclidean distance as the measure fo the distance between points and ward linkage to calculate the proximity of clusters. 

from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')

cluster.fit_predict(data)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(data[:,0], data[:, 1], c=cluster.labels_, cmap='rainbow')
```
